## Dependencies

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def demean(a):
  return a - np.mean(a) * np.array([1]*len(a))
```

```python
def correlation(a,b):
  a,b = demean(a), demean(b)
  return (a@b)/(np.linalg.norm(a)*np.linalg.norm(b))
```

```python
a = np.sin(np.linspace(0,6,100)) + np.random.uniform(-0.4,0.4,100)
b = np.sin(np.linspace(0,6,100)) + np.random.uniform(-0.4,0.4,100)
```

```python
print(f'The correlation between a and b is = ', correlation(a,b))
```

```python
fig, ax = plt.subplots(1,1,figsize=(7,7),dpi=120)
ax.plot(np.linspace(0,6,100), a, label='A series', marker='o', linestyle='dashed')
ax.plot(np.linspace(0,6,100), b, label='b series', marker='o', linestyle='dashed')
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots(1,1,figsize=(7,7),dpi=120)
ax.scatter(a,b)
ax.set_xlabel(r'$a_{i}$')
ax.set_ylabel(r'$b_{i}$')
plt.show()
```
