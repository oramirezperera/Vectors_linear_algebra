X is a vector that represents a time series of the return of investment, in percentage.
The average of the values in the time series will give us the expected percentage of return of investment and the deviation will be the risk of our investment.

## Dependencies

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = np.array([0.1]*50)
b = np.sin(np.linspace(0.4,50)) + np.random.uniform(-0.1,0.1,50)
c = np.cos(np.linspace(2,6,50)) + np.random.uniform(-0.1,0.1,50)
d = [i - np.random.uniform(0,0.8) for i in np.linspace(0,1,50)]
```

```python
fig, ax = plt.subplots(1,1,figsize=(15,7),dpi=120)
ax.plot(np.linspace(0,50,50), a, marker='o', linestyle='dashed', label='a Investment')
ax.plot(np.linspace(0,50,50), b, marker='o', linestyle='dashed', label='b Investment')
ax.plot(np.linspace(0,50,50), c, marker='o', linestyle='dashed', label='c Investment')
ax.plot(np.linspace(0,50,50), d, marker='o', linestyle='dashed', label='d Investment')
ax.set_xlabel('Time')
ax.set_ylabel('Percentage of return')
ax.legend()
plt.show()
```

```python
I = np.array([a,b,c,d])
```

```python
M = np.array([np.mean(x) for x in I])
S = np.array([np.std(x) for x in I])
```

```python
fig, ax = plt.subplots(1,1,figsize=(7,7),dpi=120)
ax.scatter(S[0], M[0], linestyle='dashed', label='a Investment')
ax.scatter(S[1], M[1], linestyle='dashed', label='b Investment')
ax.scatter(S[2], M[2], linestyle='dashed', label='c Investment')
ax.scatter(S[3], M[3], linestyle='dashed', label='d Investment')
ax.set_xlabel('Risk')
ax.set_ylabel('Expected return')
ax.legend()
plt.show()
```

```python
def z_score(x):
  z = ((x-np.mean(x))/np.std(x))
  return z
```

```python
z_a = z_score(a)
z_b = z_score(b)
z_c = z_score(c)
z_d = z_score(d)
```

```python
z_array = np.array([z_a, z_b, z_c, z_d])
```

```python
mean_z_array = np.array([np.mean(x) for x in z_array])
```

```python
fig, ax = plt.subplots(1,1,figsize=(7,7),dpi=120)
ax.plot(np.linspace(0,50,50), z_array[0], linestyle='dashed', label='a Investment')
ax.plot(np.linspace(0,50,50), z_array[1], linestyle='dashed', label='b Investment')
ax.plot(np.linspace(0,50,50), z_array[2], linestyle='dashed', label='c Investment')
ax.plot(np.linspace(0,50,50), z_array[3], linestyle='dashed', label='d Investment')
ax.set_xlabel('Time')
ax.set_ylabel('Percentage of return')
ax.legend()
plt.show()
```

```python
m_z = np.array([np.mean(x) for x in z_array])
s_z = np.array([np.std(x) for x in z_array])

fig, ax = plt.subplots(1,1,figsize=(7,7),dpi=120)
ax.scatter(s_z[0],m_z[0],linestyle='dashed',label='a investment')
ax.scatter(s_z[1],m_z[1],linestyle='dashed',label='b investment')
ax.scatter(s_z[2],m_z[2],linestyle='dashed',label='c investment')
ax.scatter(s_z[3],m_z[3],linestyle='dashed',label='d investment')

ax.set_xlabel('Risk')
ax.set_ylabel('Expected return')
ax.legend()
plt.show()
```
